```python
import os

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        import pandas as pd
        citizen = pd.read_csv(
    "/kaggle/input/datasets/merajsharif/smart-city/smart_city_citizen_activity.csv"
)

connectivity = pd.read_csv(
    "/kaggle/input/datasets/merajsharif/smart-city/urban_connectivity.csv"
)
print(citizen.shape)
print(connectivity.shape)

citizen.head()
print("Citizen Dataset Shape:", citizen.shape)
print("Connectivity Dataset Shape:", connectivity.shape)
citizen.head()
connectivity.head()
print("Citizen Columns:")
print(citizen.columns.tolist())

print("\nConnectivity Columns:")
print(connectivity.columns.tolist())
print(citizen.isnull().sum())
print(connectivity.isnull().sum())
bins = [18, 30, 45, 60, 100]
labels = ['18-29', '30-44', '45-59', '60+']

citizen['Age_Group'] = pd.cut(
    citizen['Age'],
    bins=bins,
    labels=labels,
    right=False
)

citizen[['Age', 'Age_Group']].head()
```
