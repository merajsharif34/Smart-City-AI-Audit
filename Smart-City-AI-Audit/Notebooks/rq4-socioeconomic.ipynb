```python
# =====================================================
# RQ4 - Urban Park Accessibility Across Income Groups
# =====================================================

import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import shutil

# =====================================================
# FIND DATASET
# =====================================================

connectivity_file = None

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename == "urban_connectivity.csv":
            connectivity_file = os.path.join(dirname, filename)

# =====================================================
# LOAD DATA
# =====================================================

connectivity = pd.read_csv(connectivity_file)

# =====================================================
# OUTPUT FOLDER
# =====================================================

output_dir = Path("RQ4_Output")
output_dir.mkdir(exist_ok=True)

# =====================================================
# INCOME COLUMNS
# =====================================================

income_cols = [
    "Walkable_Park_Access_low__75_city_income_",
    "Walkable_Park_Access_middle",
    "Walkable_Park_Access_high__125_city_median_income_"
]

# =====================================================
# PREPARE DATA
# =====================================================

plot_df = connectivity[income_cols].melt(
    var_name="Income_Group",
    value_name="Park_Access"
)

plot_df["Income_Group"] = plot_df["Income_Group"].replace({
    "Walkable_Park_Access_low__75_city_income_": "Low Income",
    "Walkable_Park_Access_middle": "Middle Income",
    "Walkable_Park_Access_high__125_city_median_income_": "High Income"
})

# =====================================================
# VIOLIN PLOT
# =====================================================

plt.figure(figsize=(8,5))

sns.violinplot(
    data=plot_df,
    x="Income_Group",
    y="Park_Access",
    inner="box",
    cut=0
)

plt.title(
    "Urban Park Accessibility Across Income Groups",
    fontsize=13,
    fontweight="bold"
)

plt.xlabel("Income Group")
plt.ylabel("Walkable Park Access (%)")

plt.tight_layout()

# PNG
plt.savefig(
    output_dir / "RQ4_Violin_Plot.png",
    dpi=600,
    bbox_inches="tight"
)

# PDF
plt.savefig(
    output_dir / "RQ4_Violin_Plot.pdf",
    dpi=600,
    bbox_inches="tight"
)

plt.show()

# =====================================================
# ZIP FOR DOWNLOAD
# =====================================================

shutil.make_archive(
    "RQ4_Violin_Output",
    "zip",
    output_dir
)

print("Files saved successfully:")
print("RQ4_Violin_Plot.png")
print("RQ4_Violin_Plot.pdf")
print("RQ4_Violin_Output.zip")
```
